```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
label = ['Naive Bayes', 'Support Vector Machine']
percentage = [67.55, 64.11]
fig, ax = plt.subplots(figsize=(5, 5))
index = np.arange(len(label))
index = np.arange(len(label))
ax.grid()
plt.ylim(0, 100)
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.bar(index, percentage, width=0.3)
plt.ylabel('Accuracy', fontsize=10)
plt.xticks(index, label, fontsize=10)
plt.title('Accuracy measure of each model')
plt.savefig('accuracy.png')
plt.show()
```

```python
label = ['Naive Bayes', 'Support Vector Machine']
percentage1 = [68.45, 58.08]
percentage2 = [66.83, 87.94]

fig, ax = plt.subplots(figsize=(5, 5))
index = np.arange(len(label))
index = np.arange(len(label))
ax.grid()
plt.ylim(0, 100)
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.bar(index, percentage1, width=0.3, label='Positive Precision')
plt.bar(index+0.3, percentage2, width=0.3, label='Negative Precision')
plt.legend(loc='upper right')
plt.ylabel('Precision', fontsize=10)
plt.xticks(index, label, fontsize=10)
plt.title('Precision measure of each model')
plt.savefig('precision.png')
plt.show()
```

```python
label = ['Naive Bayes', 'Support Vector Machine']
percentage1 = [62.38, 95.01]
percentage2 = [72.50, 34.69]

fig, ax = plt.subplots(figsize=(5, 5))
index = np.arange(len(label))
index = np.arange(len(label))
ax.grid()
plt.ylim(0, 100)
ttl = ax.title
ttl.set_position([.5, 1.05])
plt.bar(index, percentage1, width=0.3, label='Positive Recall', color='g')
plt.bar(index+0.3, percentage2, width=0.3, label='Negative Recall', color='r')
plt.legend(loc='upper right')
plt.ylabel('Recall', fontsize=10)
plt.xticks(index, label, fontsize=10)
plt.title('Recall measure of each model')
plt.savefig('recall.png')
plt.show()
```
